```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize

# Cargar los datos hist√≥ricos desde el archivo Excel generado previamente
data = pd.read_excel("companias_internacionales_100.xlsx")

# Configurar la columna de fechas como √≠ndice
data['Date'] = pd.to_datetime(data['Date'])
data.set_index('Date', inplace=True)

# Calcular los rendimientos diarios de cada acci√≥n
daily_returns = data.pct_change().dropna()

# Calcular la rentabilidad media diaria y la matriz de covarianza
mean_daily_returns = daily_returns.mean()
cov_matrix = daily_returns.cov()

# Funci√≥n para calcular la rentabilidad anualizada de una cartera
def annualized_return(weights):
    return np.sum(mean_daily_returns * weights) * 252

# Funci√≥n para calcular la volatilidad anualizada de una cartera
def annualized_volatility(weights):
    portfolio_variance = np.dot(weights.T, np.dot(cov_matrix, weights))
    return np.sqrt(portfolio_variance) * np.sqrt(252)

# Funci√≥n para optimizar la cartera seg√∫n una funci√≥n objetivo
def optimize_portfolio(objective_function):
    num_assets = len(mean_daily_returns)
    initial_weights = np.ones(num_assets) / num_assets  # Pesos iniciales iguales
    bounds = [(0, 1) for _ in range(num_assets)]  # Pesos entre 0% y 100%
    constraints = {'type': 'eq', 'fun': lambda weights: np.sum(weights) - 1}  # Suma de pesos = 1

    result = minimize(objective_function, initial_weights, bounds=bounds, constraints=constraints)
    return result.x  # Pesos √≥ptimos

# Funciones objetivo para cada perfil de riesgo
def risk_minimization(weights):
    return annualized_volatility(weights)

def low_risk(weights):
    return 0.75 * annualized_volatility(weights) - 0.25 * annualized_return(weights)

def medium_risk(weights):
    return 0.50 * annualized_volatility(weights) - 0.50 * annualized_return(weights)

def high_risk(weights):
    return 0.25 * annualized_volatility(weights) - 0.75 * annualized_return(weights)

def max_return(weights):
    return -annualized_return(weights)

# Optimizar las carteras para cada perfil de riesgo
profiles = {
    "Riesgo m√≠nimo": risk_minimization,
    "Riesgo bajo": low_risk,
    "Riesgo medio": medium_risk,
    "Riesgo alto": high_risk,
    "Rentabilidad m√°xima": max_return
}

results = {}
for profile, objective_function in profiles.items():
    optimal_weights = optimize_portfolio(objective_function)
    filtered_weights = {data.columns[i]: round(weight * 100, 1) for i, weight in enumerate(optimal_weights) if weight > 0.001}
    portfolio_return = annualized_return(optimal_weights)
    portfolio_volatility = annualized_volatility(optimal_weights)
    results[profile] = {
        "Composici√≥n √≥ptima": filtered_weights,
        "Rentabilidad anualizada": portfolio_return,
        "Volatilidad anualizada": portfolio_volatility
    }

    # Calcular la evoluci√≥n hist√≥rica de la cartera
    portfolio_daily_returns = (daily_returns * optimal_weights).sum(axis=1)
    portfolio_cumulative_returns = (1 + portfolio_daily_returns).cumprod()

    # Graficar la evoluci√≥n hist√≥rica
    plt.figure(figsize=(10, 6))
    plt.plot(portfolio_cumulative_returns, label=f"Cartera: {profile}")
    plt.title(f"Evoluci√≥n hist√≥rica de la cartera ({profile})")
    plt.xlabel("Fecha")
    plt.ylabel("Crecimiento acumulado")
    plt.legend()
    plt.grid()
    plt.show()

# Mostrar los resultados
for profile, result in results.items():
    print(f"\nPerfil: {profile}")
    print("Composici√≥n √≥ptima de la cartera:")
    for company, weight in result["Composici√≥n √≥ptima"].items():
        print(f"  {company}: {weight}%")
    print(f"Rentabilidad anualizada: {result['Rentabilidad anualizada']:.2%}")
    print(f"Volatilidad anualizada: {result['Volatilidad anualizada']:.2%}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
from scipy.optimize import minimize

# --- Carga y preparaci√≥n de datos ---
data = pd.read_excel("companias_internacionales_100.xlsx")
data['Date'] = pd.to_datetime(data['Date'])
data.set_index('Date', inplace=True)
daily_returns  = data.pct_change().dropna()
mean_daily_ret = daily_returns.mean()
cov_matrix     = daily_returns.cov()

# --- Funciones de m√©tricas ---
def annualized_return(w):
    return np.sum(mean_daily_ret * w) * 252

def annualized_volatility(w):
    var = np.dot(w.T, np.dot(cov_matrix, w))
    return np.sqrt(var) * np.sqrt(252)

# --- Funci√≥n de optimizaci√≥n gen√©rica ---
def optimize_portfolio(obj_fn):
    n    = len(mean_daily_ret)
    w0   = np.ones(n) / n
    bds  = [(0, 1)] * n
    cons = {'type': 'eq', 'fun': lambda w: np.sum(w) - 1}
    res  = minimize(obj_fn, w0, bounds=bds, constraints=cons)
    return res.x

# --- Objetivos por perfil ---
profiles = {
    "Riesgo m√≠nimo":       annualized_volatility,
    "Riesgo bajo":         lambda w: 0.75*annualized_volatility(w) - 0.25*annualized_return(w),
    "Riesgo medio":        lambda w: 0.50*annualized_volatility(w) - 0.50*annualized_return(w),
    "Riesgo alto":         lambda w: 0.25*annualized_volatility(w) - 0.75*annualized_return(w),
    "Rentabilidad m√°xima": lambda w: -annualized_return(w)
}

# --- Ejecutar optimizaciones y recolectar resultados ---
results = {}
os.makedirs("data", exist_ok=True)

for name, fn in profiles.items():
    w_opt = optimize_portfolio(fn)
    ret   = annualized_return(w_opt)
    vol   = annualized_volatility(w_opt)

    # DataFrame de composici√≥n
    dfw = pd.DataFrame({
        'Activo':   data.columns,
        'Peso (%)': np.round(w_opt * 100, 2)
    }).query("`Peso (%)` > 0.1").reset_index(drop=True)

    results[name] = {'df_weights': dfw, 'return': ret, 'volatility': vol}

    # Gr√°fica de evoluci√≥n hist√≥rica con mayor resoluci√≥n
    portfolio_daily_returns    = (daily_returns * w_opt).sum(axis=1)
    portfolio_cum_returns      = (1 + portfolio_daily_returns).cumprod()

    plt.figure(figsize=(10, 6), dpi=150)
    plt.plot(portfolio_cum_returns, label=f"Cartera: {name}")
    plt.title(f"Evoluci√≥n hist√≥rica de la cartera ({name})")
    plt.xlabel("Fecha")
    plt.ylabel("Crecimiento acumulado")
    plt.legend()
    plt.grid()
    plt.show()

# --- Escritura en Excel con formato optimizado ---
import os

# --- Ruta de salida personalizada en Windows ---
out_dir  = r"C:\Users\marco\Desktop\Proyecto_Final\data"
os.makedirs(out_dir, exist_ok=True)

out_path = os.path.join(out_dir, "resultados_carteras_optimizado.xlsx")

# Si existe, lo borramos para poder sobreescribirlo
if os.path.exists(out_path):
    try:
        os.remove(out_path)
    except PermissionError:
        raise PermissionError(f"No se pudo eliminar {out_path}. ¬øLo tienes abierto en Excel?")

with pd.ExcelWriter(out_path, engine="xlsxwriter") as writer:
    workbook = writer.book
    fmt_pct  = workbook.add_format({'num_format': '0.00%', 'align':'right'})
    fmt_num  = workbook.add_format({'num_format': '0.00', 'align':'right'})
    fmt_txt  = workbook.add_format({'bold': True})

    # Hoja Resumen
    resumen = pd.DataFrame([
        {
            'Perfil':                   name,
            'Rentabilidad anualizada':  r['return'],
            'Volatilidad anualizada':   r['volatility']
        }
        for name, r in results.items()
    ])
    resumen.to_excel(writer, sheet_name="Resumen", index=False)
    ws = writer.sheets["Resumen"]
    ws.set_column("A:A", 18)
    ws.set_column("B:C", 22, fmt_pct)

    # Hojas por perfil
    for name, r in results.items():
        sheet = name[:31]
        r['df_weights'].to_excel(writer, sheet_name=sheet, index=False)
        ws = writer.sheets[sheet]
        ws.set_column("A:A", 20, fmt_txt)
        ws.set_column("B:B", 12, fmt_num)
        # M√©tricas al pie
        end_row = len(r['df_weights']) + 2
        ws.write(f"A{end_row}",     "Rentabilidad anualizada", fmt_txt)
        ws.write(f"B{end_row}",     r['return'],              fmt_pct)
        ws.write(f"A{end_row+1}",   "Volatilidad anualizada",  fmt_txt)
        ws.write(f"B{end_row+1}",   r['volatility'],           fmt_pct)

print(f"\n‚úÖ Archivo Excel generado en: {out_path}\n")

# --- Impresi√≥n de resultados en consola (valores detallados) ---
for name, r in results.items():
    print(f"Perfil: {name}")
    print("  Composici√≥n √≥ptima:")
    for _, row in r['df_weights'].iterrows():
        print(f"    {row['Activo']}: {row['Peso (%)']}%")
    print(f"  Rentabilidad anualizada: {r['return']:.2%}")
    print(f"  Volatilidad anualizada: {r['volatility']:.2%}\n")


```
